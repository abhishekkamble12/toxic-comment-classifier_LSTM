```python
vectorizer = TextVectorization(max_tokens=max_features,
                               output_sequence_length=1800,
                               output_mode='int')
```

Tensor flow datapipeline

```python
# 1. Create the base dataset
dataset = tf.data.Dataset.from_tensor_slices((vectorized_text, y))

# 2. Apply caching, shuffling, BATCHING, and prefetching to the SAME variable
# (Note: We use 'dataset' consistently here to avoid losing the batching step)
dataset = dataset.cache()
dataset = dataset.shuffle(160000)
dataset = dataset.batch(16)  # <--- This adds the 3rd dimension your model needs
dataset = dataset.prefetch(8)

# 3. Create your splits from the BATCHED dataset
train = dataset.take(int(len(dataset)*.7))
val = dataset.skip(int(len(dataset)*.7)).take(int(len(dataset)*.2))
test = dataset.skip(int(len(dataset)*.9)).take(int(len(dataset)*.1))

# 4. Verify the shape (Should be (Batch_Size, Sequence_Length))
batch_X, batch_y = train.as_numpy_iterator().next()
print("Batch shape:", batch_X.shape)
# Expected Output: (16, 1800) -> This 16 is the missing dimension!
```

DEEP MODEL BUILDING

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Bidirectional, Dense, Embedding
```

```python
model = Sequential()
model.add(Embedding(max_features+1, 32))
model.add(Bidirectional(LSTM(32, activation='tanh')))
model.add(Dense(128, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(6, activation='sigmoid'))
#
```

```python
model.compile(loss='BinaryCrossentropy', optimizer='Adam')
```

```python
# The shape is (Batch_Size, Sequence_Length)
# None allows for any batch size, and 1800 is your text length
model.build(input_shape=(None, 1800))

model.summary()
```

```python
history = model.fit(train, epochs=1, validation_data=val)
```

```python
# !pip install gradio

import gradio as gr
import tensorflow as tf
import numpy as np

# Load the model (if you haven't already in the specific cell above)
model = tf.keras.models.load_model('toxic_comment_model.h5')

def score_comment(comment):
    # 1. Vectorize the input comment using the vectorizer you created earlier
    vectorized_comment = vectorizer([comment])

    # 2. Predict using the loaded model
    results = model.predict(vectorized_comment)

    # 3. Create a formatted output dictionary for the 6 classes
    # The columns in order based on your dataset
    columns = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']

    # Gradio's Label output expects a dictionary of {label: probability}
    output_dict = {}
    for i, col in enumerate(columns):
        output_dict[col] = float(results[0][i])

    return output_dict

# Create the Gradio Interface
interface = gr.Interface(
    fn=score_comment,
    inputs=gr.Textbox(lines=2, placeholder="Enter a comment..."),
    outputs=gr.Label(num_top_classes=6),
    title="Toxic Comment Classifier",
    description="Enter a comment to see the probability of each toxicity category."
)

# Launch the app inline in the notebook
interface.launch(share=True)
```
